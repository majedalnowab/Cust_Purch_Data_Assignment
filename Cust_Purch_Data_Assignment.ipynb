**4. What are the max and min ages of your customer? Can you find mean of your customer?**

```python
# Code here please

# Get the max age
max_age = df["age"].max()

# Get the min age
min_age = df["age"].min()

# Get the mean age
mean_age = df["age"].mean()

print(f"Max. age of the customer: {max_age}")
print(f"Min. age of the customer: {min_age}")
print(f"Avg. age of the customer: {mean_age}")
```

**5. What are the three most common customer's names?**<br>
&#9989; [<code>**value_counts()**</code>](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) returns object containing counts of unique values. The resulting object will be in descending order so that the first element is the most frequently-occurring element. Excludes NA values by default.

```python
# Code here please

# Get the top 3 most common names and counts
top_3_names_and_counts = df["first"].value_counts().head(3)

# Format the output using f-string
output = "\n".join(f"{name}: {count}" for name, count in top_3_names_and_counts.items())

# Print the output
print(output)
```

**10. What is the max, min and average spending?**

```python
# Code here please

# Calculate max, min, and average spending
max_spending = df['price(CAD)'].max()
min_spending = df['price(CAD)'].min()
average_spending = df['price(CAD)'].mean()

# Print or use the results as needed
print(f"Maximum Spending: {max_spending}")
print(f"Minimum Spending: {min_spending}")
print(f"Average Spending: {average_spending}")
```

**12. As a loyalty reward, company wants to send thanks coupon to those who spent 100CAD or more, please find out the customers?**

**13. How many emails are associated with this credit card number '5020000000000230'?**

```python
# Code here please

# Credit card number to search for
target_credit_card = 5020000000000230

# Find emails associated with the specified credit card number
matching_emails = df[df['cc_no'] == target_credit_card]['email']

# Number of emails and the corresponding email addresses
num_emails = len(matching_emails)
email_addresses = matching_emails.tolist()

# Print the results
print(f"Number of Emails: {num_emails}")
print(f"Email Addresses: {email_addresses}")
```

**14. We need to send new cards to the customers well before the expire, how many cards are expiring in 2019?**<br>
*Use `sum()` and `count()` and see the difference in their use :)*

```python
# Code here please

# Convert cc_exp to datetime
df['cc_exp'] = pd.to_datetime(df['cc_exp'], format='%m/%Y')

# Filter cards expiring in 2019
expiring_in_2019 = df[df['cc_exp'].dt.year == 2019]

# Using sum() to count the number of cards
num_cards_sum = expiring_in_2019['cc_no'].count()

# Using count() to count the number of rows
num_cards_count = expiring_in_2019['cc_no'].count()

# Print the results
print(f"Number of Cards (Using sum()): {num_cards_sum}")
print(f"Number of Cards (Using count()): {num_cards_count}")
```

**17. What are two most common professions?**

```python
# Code here please

# Find the two most common professions
top_professions = df['profession'].value_counts().nlargest(2)

# Print the result
print(top_professions)
```

**18. Can you tell the top 5 most popular email providers? (e.g. gmail.com, yahoo.com, etc...)**

```python
# Code here please

# Extract email domains and count occurrences
email_domains = df['email'].str.split('@').str[1]
top_email_providers = email_domains.value_counts().nlargest(5)

# Print the result
print(top_email_providers)
```

**19. Is there any customer who is using email with "am.edu"?**<br>
Hint: Use `lambda` expression in `apply()`. split the email address at `@`.  

```python
# Code here please

# Find the customer using email with "am.edu"
target_customer = df[df['email'].apply(lambda x: x.split('@')[1] == 'am.edu')]

target_customer
```
